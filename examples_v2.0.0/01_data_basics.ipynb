This notebook shows how to load and visualize two benchmark datasets -- STEAD and INSTANCE dataset.

##  $(1)$ STEAD Dataset

### $(\mathrm{i})$ Loading the STEAD dataset
Please make sure to download the dataset following the instructions given in https://github.com/smousavi05/STEAD and save it to the desired storage path in a folder named STEAD.

```python
#import sys
#sys.path.insert(0, '..')
from saipy.data.base import STEAD
stead = STEAD(directory = '/home/seismoai/data')
print(stead)
```

```python
stead.metadata.head()
```

### $(\mathrm{ii})$ Getting a list of traces

```python
stead.trace_list()
```

### $(\mathrm{iii})$ Checking distribution of parameters
Below, we can generate plots depicting various distributions, such as earthquake magnitude, source distance (km), signal-to-noise (SNR) ratio (dB), etc.

```python
stead.distribution('source_magnitude', log = True)
```

```python
stead.distribution('source_distance_km', log = True, color = 'salmon')
```

```python
stead.distribution('snr_db')
```

```python
stead.distribution('source_longitude')
```

## (2) INSTANCE Dataset

###   $(\mathrm{i})$ Loading the INSTANCE dataset
Please make sure to download the dataset following the instructions given in https://github.com/INGV/instance and save it to the desired storage path in a folder named INSTANCE.

```python
from saipy.data.base import INSTANCE
instance = INSTANCE(directory = '/home/seismoai/data')
print(instance)
```

### $(\mathrm{ii})$ Getting a list of traces¶

earthquake traces

```python
instance.trace_list_events()
```

noise traces

```python
instance.trace_list_noise()
```

### $(\mathrm{iii})$ Checking distribution of parameters

```python
instance.distribution('source_magnitude', log = True)
```

```python
instance.distribution('path_ep_distance_km', log = True, color = 'salmon')
```

```python
instance.distribution('station_longitude_deg')
```
