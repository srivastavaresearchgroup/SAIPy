# Applying SAIPy: Downloading data from IRIS - single station

This notebook demonstrates the use of the package for event detection, magnitude estimation, phase arrival-time picking and polarity estimation, and saving the result as a csv file.
* Updated in May 2025.

```python
import os
import torch
import obspy
from obspy.clients.fdsn import Client
from obspy import UTCDateTime
from saipy.data.realdata import *
from saipy.utils.packagetools import *
```

```python
%load_ext autoreload
%autoreload 2
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

### $(\mathrm{i})$ Path to the Models

```python
dirmodel = '../saipy/saved_models/'
```

### $(\mathrm{ii})$ Specifying waveform details

```python
wsp = 'IRIS'
network =  'BC'
station =  'VTX'
location = '*'
channel = "HH*"
start_time = UTCDateTime("2023-05-02T06:30:00")
end_time = UTCDateTime("2023-05-02T07:30:00")
```

### $(\mathrm{iii})$ Monitoring continuous data. Last Version v2 (monitor function).

```python
# Time format for titles
starttime_str = start_time.strftime("%Y-%m-%d %H:%M:%S")
day = start_time.strftime("%Y%m%d")
stime = start_time.strftime("%H%M%S")
```

```python
print(f"\n** Downloading data: station {station}, date {starttime_str}... **")
stream = waveform_download(wsp, net=network, sta=station, loc=location, chan=channel, starttime=start_time, endtime=end_time)
print(stream)
```

```python
# Filter bp 1-45 Hz and resampling (100 Hz, if necessary)
print("** Preprocessing data... **")
prepro_stream, prepro_wave_array = preprocessing(stream)
print(prepro_stream)
```

```python
# Visualization

plot_range = np.arange(0, len(prepro_wave_array))
time_utc = start_time.datetime
fig = plot_data(prepro_wave_array, station, time_utc, starttime_str, samples=plot_range)
```

```python
# Call ** monitor ** process
lw = 30
dw = 5
output_path = '../results/results_IRIS_v2' # edit

outputs = monitor(prepro_stream, prepro_wave_array, device,
                                  station, day, stime,
                                  len_win=lw, detection_windows=dw,
                                  model_path=dirmodel,
                                  save_result=True,
                                  outpath=output_path)
```

### $(\mathrm{iv})$ Monitoring continuous data. Version v1 (monitor1 function).
