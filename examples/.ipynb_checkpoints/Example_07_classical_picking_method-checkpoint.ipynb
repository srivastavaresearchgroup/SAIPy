This tutorial shows how to use STA/LTA method (e.g., classical and recursive STA/LTA) for phase pikcing.

### read trace data

```python
import obspy
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
from obspy.core import read
from obspy.signal.trigger import plot_trigger
trace = read("https://examples.obspy.org/ev0_6.a01.gse2")[0]
trace.plot();
```

```python
from deepseis.utils.packagetools import *
cft = class_pick(trigger_type = 'classic_sta_lta', 
                 trace = trace, nsta=5, nlta=10, 
                 thr_on=1.5, thr_off=0.5, plotFlag=True)
```

### direclt stream data

```python
from obspy.clients.fdsn import Client
# Retrieve waveforms via FDSNWS
client = Client("LMU")
t = obspy.UTCDateTime("2009-08-24 00:19:45")
st = client.get_waveforms('BW', 'RTSH', '', 'EHZ', t, t + 50)
st.plot();
```

```python
from deepseis.utils.packagetools import *
cft = class_pick(trigger_type='recursive_sta_lta', trace =st[0], 
                 nsta=2.5, nlta=10, thr_on=3.5, thr_off=0.5, 
                 plotFlag=True)
```
